```python
from sklearn.preprocessing import LabelEncoder
import joblib

# Create a dictionary to store label encoders
label_encoders = {}

categorical_cols = ['workclass', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']

for col in categorical_cols:
    encoder = LabelEncoder()
    data[col] = encoder.fit_transform(data[col])
    label_encoders[col] = encoder # Store the fitted encoder

# Save all fitted label encoders
joblib.dump(label_encoders, 'label_encoders.pkl')
print("✅ Saved label encoders as label_encoders.pkl")

```

```python
data
```

```python
x=data.drop(columns=['income'])
y=data['income']
x
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
import joblib

# Ensure 'income' is encoded numerically for consistent target variable
# LabelEncoder is suitable for binary classification like income <=50K vs >50K.
income_encoder = LabelEncoder()
y_encoded = income_encoder.fit_transform(y)
joblib.dump(income_encoder, 'income_encoder.pkl')
print("✅ Saved income encoder for target variable as income_encoder.pkl")

X_train, X_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42)

models = {
    "LogisticRegression": LogisticRegression(max_iter=1000, random_state=42), # Added random_state
    "RandomForest": RandomForestClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(random_state=42),
    "GradientBoosting": GradientBoostingClassifier(random_state=42)
}

results = {}
pipelines = {} # To store pipelines

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()), # Step 1: Scale numerical features
        ('model', model)             # Step 2: Train the model
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    pipelines[name] = pipe # Store the entire pipeline
    print(f"\n{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))

# Get best model and its pipeline
best_model_name = max(results, key=results.get)
best_pipeline = pipelines[best_model_name]
print(f"\n✅ Best model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save the best model (which is now a pipeline including the scaler)
joblib.dump(best_pipeline, "best_model_pipeline.pkl")
print("✅ Saved best model pipeline as best_model_pipeline.pkl")
```

```python
import matplotlib.pyplot as plt
plt.bar(results.keys(), results.values(), color='skyblue')
plt.ylabel('Accuracy Score')
plt.title('Model Comparison')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import numpy as np # Import numpy for NaN handling

# --- Load Model and Encoders ---
try:
    model_pipeline = joblib.load("best_model_pipeline.pkl")
    label_encoders = joblib.load("label_encoders.pkl")
    income_encoder = joblib.load("income_encoder.pkl") # To decode predictions
except FileNotFoundError:
    st.error("Error: Model or encoders not found. Please run the Jupyter notebook cells to train and save them.")
    st.stop() # Stop the app if files are not found

# --- Page Configuration ---
st.set_page_config(
    page_title="Employee Salary Classification App 💼",
    page_icon="💸",
    layout="wide", # Use 'wide' layout for better spacing
    initial_sidebar_state="expanded"
)

# --- Custom CSS for Styling ---
st.markdown("""
<style>
    .stApp {
        background-color: #f0f2f6; /* Light gray background */
        color: #333333; /* Darker text */
    }
    .stButton>button {
        background-color: #4CAF50; /* Green button */
        color: white;
        border-radius: 5px;
        border: none;
        padding: 10px 20px;
        font-size: 16px;
        cursor: pointer;
        box-shadow: 2px 2px 5px rgba(0,0,0,0.2); /* Subtle shadow */
        transition: all 0.3s ease; /* Smooth transition */
    }
    .stButton>button:hover {
        background-color: #45a049; /* Darker green on hover */
        box-shadow: 3px 3px 8px rgba(0,0,0,0.3); /* Larger shadow on hover */
        transform: translateY(-2px); /* Slight lift */
    }
    .stTextInput>div>div>input, .stNumberInput>div>div>input {
        border-radius: 5px;
        border: 1px solid #ccc;
        padding: 8px;
        box-shadow: inset 1px 1px 3px rgba(0,0,0,0.1); /* Inner shadow for depth */
    }
    .stSelectbox>div>div {
        border-radius: 5px;
        border: 1px solid #ccc;
        padding: 4px; /* Adjust padding for selectbox */
        box-shadow: inset 1px 1px 3px rgba(0,0,0,0.1);
    }
    h1 {
        color: #2c3e50; /* Dark blue/gray for main title */
        text-align: center;
        font-size: 2.8em; /* Slightly larger */
        margin-bottom: 25px;
        font-family: 'Segoe UI', sans-serif; /* Modern font */
        text-shadow: 1px 1px 2px rgba(0,0,0,0.1);
    }
    h2 {
        color: #34495e; /* Slightly lighter for subheadings */
        font-size: 2em;
        border-bottom: 2px solid #aec6cf; /* Subtle underline */
        padding-bottom: 5px;
        margin-top: 30px;
        margin-bottom: 20px;
    }
    h3 {
        color: #34495e;
        font-size: 1.6em;
        margin-bottom: 15px;
    }
    .stAlert {
        border-radius: 8px;
    }
    .prediction-result {
        background-color: #e6ffe6; /* Light green background for success */
        border-left: 6px solid #4CAF50; /* Thicker green border */
        padding: 20px;
        margin-top: 30px;
        border-radius: 10px;
        font-size: 1.8em; /* Larger font for result */
        font-weight: bold;
        text-align: center;
        color: #28a745; /* Darker green text */
        box-shadow: 0px 4px 10px rgba(0,0,0,0.15); /* More prominent shadow */
    }
    .stMarkdown p {
        font-size: 1.15em; /* Slightly larger text */
        line-height: 1.7;
        color: #444;
    }
    /* Customize sidebar */
    .st-emotion-cache-1ldf5b-container {
        background-color: #f7f9fc; /* Off-white for sidebar */
        padding: 25px;
        border-right: 1px solid #e0e0e0;
        box-shadow: 2px 0px 5px rgba(0,0,0,0.05);
    }
    .st-emotion-cache-1ldf5b-container .st-emotion-cache-10q7pck {
        font-size: 1.6em; /* Increase sidebar header font size */
        font-weight: bold;
        color: #2c3e50;
        margin-bottom: 20px;
        text-shadow: 0.5px 0.5px 1px rgba(0,0,0,0.1);
    }
    .st-emotion-cache-1ldf5b-container .stSlider label,
    .st-emotion-cache-1ldf5b-container .stSelectbox label,
    .st-emotion-cache-1ldf5b-container .stNumberInput label {
        font-size: 1.1em;
        font-weight: 500;
        color: #333333;
        margin-bottom: 5px;
    }
    .block-container {
        padding-top: 2rem;
        padding-bottom: 2rem;
    }
</style>
""", unsafe_allow_html=True)

# --- Main Title and Description ---
st.title("💸 Employee Salary Classifier")
st.markdown("""
<p style='font-size:1.2em; text-align: center; color: #555;'>
Predict whether an employee's income is <span style='font-weight:bold;'>&gt;50K</span> or <span style='font-weight:bold;'>≤50K</span>
based on their demographic and employment details.
</p>
""", unsafe_allow_html=True)

# --- Input Section ---
st.header("⚙️ Employee Details for Prediction")
st.markdown("Please provide the following information to get a salary class prediction.")

# Use columns for a better layout of input fields
col1, col2, col3 = st.columns(3)

with col1:
    st.markdown("### Personal & Work Info")
    age = st.slider("Age", 17, 75, 30) # Adjusted min/max based on your outlier removal
    workclass_options = label_encoders['workclass'].classes_
    workclass = st.selectbox("Workclass", workclass_options)
    gender_options = label_encoders['gender'].classes_
    gender = st.selectbox("Gender", gender_options)
    race_options = label_encoders['race'].classes_
    race = st.selectbox("Race", race_options)

with col2:
    st.markdown("### Education & Marital Status")
    # Mapping education level to educational-num based on your dataset's typical values
    # These are derived from observations of the 'education' and 'educational-num' columns in 'adult 3.csv'
    education_to_num = {
        "Preschool": 1, "1st-4th": 2, "5th-6th": 3, "7th-8th": 4, "9th": 5, "10th": 6,
        "11th": 7, "12th": 8, "HS-grad": 9, "Some-college": 10, "Assoc-voc": 11,
        "Assoc-acdm": 12, "Bachelors": 13, "Masters": 14, "Prof-school": 15, "Doctorate": 16
    }
    # Ensure options are displayed in a human-readable order
    education_display_options = sorted(education_to_num.keys(), key=lambda x: education_to_num[x])
    selected_education_str = st.selectbox("Education Level", education_display_options, index=education_display_options.index("Bachelors")) # Default to Bachelors
    educational_num = education_to_num[selected_education_str] # Get the numerical value for the model

    marital_status_options = label_encoders['marital-status'].classes_
    marital_status = st.selectbox("Marital Status", marital_status_options)
    relationship_options = label_encoders['relationship'].classes_
    relationship = st.selectbox("Relationship", relationship_options)

with col3:
    st.markdown("### Financial & Other Details")
    occupation_options = label_encoders['occupation'].classes_
    occupation = st.selectbox("Occupation", occupation_options)
    hours_per_week = st.slider("Hours per Week", 1, 80, 40) # Adjusted min/max based on your outlier removal
    capital_gain = st.number_input("Capital Gain", min_value=0, max_value=100000, value=0, help="Amount of capital gains (e.g., from investments)")
    capital_loss = st.number_input("Capital Loss", min_value=0, max_value=100000, value=0, help="Amount of capital losses")
    fnlwgt = st.number_input("Fnlwgt (Final Weight)", min_value=10000, max_value=1000000, value=200000, help="The number of people the census believes the entry represents") # Placeholder range
    native_country_options = label_encoders['native-country'].classes_
    native_country = st.selectbox("Native Country", native_country_options, index=list(native_country_options).index('United-States'))

# --- Prepare Input Data for Prediction ---
# Create a dictionary matching the original DataFrame's columns and their order
# This order is CRUCIAL and must match the 'x' DataFrame fed to your model in the notebook.
# You can get this exact order by printing X_train.columns after Cell 32 runs.
# Based on your notebook's processing, this should be the order:
expected_columns_order = [
    'age', 'workclass', 'fnlwgt', 'educational-num', 'marital-status', 'occupation',
    'relationship', 'race', 'gender', 'capital-gain', 'capital-loss',
    'hours-per-week', 'native-country'
]

input_data_dict = {
    'age': age,
    'workclass': workclass,
    'fnlwgt': fnlwgt,
    'educational-num': educational_num,
    'marital-status': marital_status,
    'occupation': occupation,
    'relationship': relationship,
    'race': race,
    'gender': gender,
    'capital-gain': capital_gain,
    'capital-loss': capital_loss,
    'hours-per-week': hours_per_week,
    'native_country': native_country # This needs to match the column name from your data, which is 'native-country'
}

# Adjust key to match exact column name from your dataset
input_data_dict['native-country'] = input_data_dict.pop('native_country')


# Create a DataFrame from the input dictionary
input_df_raw = pd.DataFrame([input_data_dict])

# Display raw input data
st.markdown("---")
st.markdown("### 📊 Raw Input Data Provided")
st.dataframe(input_df_raw)

# --- Preprocess Input Data (Mimic Notebook's Preprocessing) ---
def preprocess_input(df_raw, encoders):
    df = df_raw.copy()

    # Apply Label Encoding using the loaded encoders
    for col, encoder in encoders.items():
        # Handle unseen labels: if a category in the input is not in the encoder's classes_,
        # we will map it to the most frequent class during training (class with index 0, often 'Private' or 'Husband')
        # This is a robust way to handle new, unexpected categories.
        # However, with Streamlit selectboxes, this is largely a safeguard as options are predefined.
        if col in df.columns:
            # Create a mapping for known classes, and a default for unknown
            mapping_dict = {cls: idx for idx, cls in enumerate(encoder.classes_)}
            # Apply mapping, if a value is not in mapping_dict, use a default (e.g., 0, or an average)
            # For simplicity, we assume predefined selectbox options prevent unseen values.
            df[col] = df[col].map(mapping_dict)
            # Fill any NaN values that resulted from unseen categories with a default (e.g., 0)
            df[col] = df[col].fillna(0).astype(int) # Default to 0 if an issue occurred
        else:
            st.warning(f"Column '{col}' expected for encoding but not found in input data. Skipping encoding for this column.")

    # Ensure all columns are numeric after encoding
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        if df[col].isnull().any():
            st.error(f"Error converting column '{col}' to numeric. Please check inputs.")
            st.stop()
    return df

processed_input_df = preprocess_input(input_df_raw, label_encoders)

# Ensure column order matches the training data's X_train.columns (from your notebook)
processed_input_df = processed_input_df[expected_columns_order]

st.markdown("### ✨ Processed Input Data (Ready for Model)")
st.dataframe(processed_input_df)

# --- Prediction Section ---
st.markdown("---")
if st.button("🚀 Predict Salary Class"):
    try:
        # Make prediction using the loaded pipeline (which includes scaling)
        prediction_numeric = model_pipeline.predict(processed_input_df)[0]
        prediction_class = income_encoder.inverse_transform([prediction_numeric])[0]

        st.markdown(f"""
        <div class="prediction-result">
            Predicted Income Class: {prediction_class}
        </div>
        """, unsafe_allow_html=True)
    except Exception as e:
        st.error(f"An error occurred during prediction: {e}")
        st.exception(e) # Display full exception for debugging

# --- Batch Prediction Section ---
st.markdown("---")
st.markdown("## 📁 Batch Prediction (Upload CSV)")
st.markdown("Upload a CSV file containing employee data (with the same columns as the input form) to get predictions for multiple entries.")

uploaded_file = st.file_uploader("Choose a CSV file", type="csv")

if uploaded_file is not None:
    try:
        batch_data_raw = pd.read_csv(uploaded_file)
        st.markdown("### Uploaded Data Preview:")
        st.dataframe(batch_data_raw.head())

        # Preprocess batch data
        batch_processed_df = preprocess_input(batch_data_raw.copy(), label_encoders) # Use copy to avoid modifying raw data preview
        batch_processed_df = batch_processed_df[expected_columns_order] # Ensure column order

        # Make predictions
        batch_predictions_numeric = model_pipeline.predict(batch_processed_df)
        batch_predictions_class = income_encoder.inverse_transform(batch_predictions_numeric)

        batch_data_raw['Predicted_Income_Class'] = batch_predictions_class

        st.markdown("### ✅ Predictions for Batch Data:")
        st.dataframe(batch_data_raw.head())

        csv = batch_data_raw.to_csv(index=False).encode('utf-8')
        st.download_button(
            label="⬇️ Download Predictions CSV",
            data=csv,
            file_name="predicted_employee_salaries.csv",
            mime="text/csv",
        )
    except Exception as e:
        st.error(f"An error occurred during batch prediction: {e}")
        st.exception(e) # Display full exception for debugging

st.markdown("---")
st.info("Developed with ❤️ for IBM SkillsBuild Project")
```
